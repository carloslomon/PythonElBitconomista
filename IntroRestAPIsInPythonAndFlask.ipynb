# Key Differences
- __str__: End users who need a readable and informative description of the object
- __repr__: Developers who need an unambiguous representation of the object, which can often be used to recreate the object.

Default Behavior:

If __str__ is not defined, Python will use __repr__ as a fallback for str() and print().
If neither __str__ nor __repr__ is defined, the default implementation inherited from object is used, which typically includes the object's type and memory address.


```python
class MyClass:
    def __init__(self, value):
        self.value = value

    def __str__(self):
        return f"MyClass with value: {self.value}"

    def __repr__(self):
        return f"MyClass({self.value})"

obj = MyClass(10)
print(str(obj))  # Output: MyClass with value: 10
print(repr(obj)) # Output: MyClass(10)
print(obj)       # Output: MyClass with value: 10 (since print() uses str() under the hood)
```

```python
# We will develop two new classes
# Here, we will encounter an error of inheritance used very clumsily
# What is the major error here? Hint ith is an inheritance Error
class CarCarrierTrailer:
  def __init__(self, trailer_name):
    self.number_of_cars = []
    self.trailer_name = trailer_name
  def __str__(self):
    if self.number_of_cars is not None:
      return f"{self.trailer_name} is the car carrier trailer that has {len(self.number_of_cars)} cars"
    else:
      return f"{self.trailer_name} may not be a trailer car"

  def add_car(self, car):
    self.number_of_cars.append(car)

class Car(CarCarrierTrailer):
  """

  Attributes:
    model:
    year:
    color:
  """
  def __init__(self, model, year, color, mileage):
    self.model = model
    self.year = year
    self.color = color
    self.mileage = mileage
    super().__init__(model)
  def __str__(self):
    if self.model == "Porsche-911" and self.color != "red":
      self.color = "red"
      super().__str__()
    else:
      return f"The {self.color} {self.model} is from the year {self.year} and has a mileage of {self.mileage}"
```

```python
# What is the error here
class Book:
    def __init__(self, title: str , author: str, paper_price_dollars: float, paper_weight_kg: float, btype="Coverless"):
        self.title = title
        self.author = author
        self.price_dollars = paper_price_dollars
        self.weight_kg = paper_weight_kg
        self.btype = btype
    def __repr__(self) -> str:
        return f"{self.btype} book ({self.title}, {self.author}, ${self.price_dollars}, {self.weight_kg} kg)"
    @classmethod
    def hardcover(cls, title: str, author: str, price: float, weight: float)->Book:
      weight_kg = 0.4 + weight
      price_dollars_n = 7.5 + price
      return cls(title, author, price_dollars_n, weight_kg, "Hrdcover")
    @classmethod
    def paperback(cls, title: str, author: str, price: float, weight: float)->Book:
      weight_kg = 0.1 + weight
      price_dollars_n = 2.5 + price
      return cls(title, author, price_dollars_n, weight_kg, "Paperback")

b1 = Book("Harry Potter and the Half-Blood Prince", "J.K. Rowling", 10.5, 0.85)
print(b1)
print(b1.hardcover(b1.title, b1.author, b1.price_dollars, b1.weight_kg))
print(b1.paperback(b1.title, b1.author, b1.price_dollars, b1.weight_kg))

"""By removing the @classmethod decorator, hardcover and paperback become instance methods that can correctly modify the attributes of the book instance (b1 in this case)."""
```
